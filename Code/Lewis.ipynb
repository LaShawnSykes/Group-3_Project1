```python
import pandas as pd

# Load the data from the provided Excel file
file_path = 'clean_summary_data_Mean_charts.xlsx'
data = pd.read_excel(file_path)

# Display the first few rows of the dataframe to understand its structure
data.head()
```

```python
from scipy.stats import ttest_ind, f_oneway, chi2_contingency
import numpy as np

# Preparing data for statistical tests
# T-test for BMI based on Gender
male_bmi = data[data['Gender (Female=1, Male=2)'] == 2]['BMI (kg/m2)']
female_bmi = data[data['Gender (Female=1, Male=2)'] == 1]['BMI (kg/m2)']

# ANOVA for BMI based on Type of Diabetes
types_diabetes = data['Type of Diabetes'].unique()
groups_bmi = [data[data['Type of Diabetes'] == diabetes]['BMI (kg/m2)'] for diabetes in types_diabetes]

# Chi-Square Test for association between Gender and Type of Diabetes
contingency_table = pd.crosstab(data['Gender (Female=1, Male=2)'], data['Type of Diabetes'])

# Performing T-test
t_test_results = ttest_ind(male_bmi.dropna(), female_bmi.dropna())

# Performing ANOVA
anova_results = f_oneway(*[group.dropna() for group in groups_bmi])

# Performing Chi-Square Test
chi_square_results = chi2_contingency(contingency_table)

t_test_results, anova_results, chi_square_results
```

```python
# Checking the distribution of data for 'Type of Diabetes' to understand the ANOVA issues
data['Type of Diabetes'].value_counts()
```

```python
# Check for missing values in BMI for both types of diabetes
missing_bmi_t1dm = data[data['Type of Diabetes'] == 'T1DM']['BMI (kg/m2)'].isna().sum()
missing_bmi_t2dm = data[data['Type of Diabetes'] == 'T2DM']['BMI (kg/m2)'].isna().sum()

# Verify that there are no missing values
missing_bmi_t1dm, missing_bmi_t2dm
```

```python
# Redo ANOVA for BMI based on Type of Diabetes
# As we have verified no NaN values in the BMI data, we can confidently proceed with the ANOVA test
t1dm_bmi = data[data['Type of Diabetes'] == 'T1DM']['BMI (kg/m2)']
t2dm_bmi = data[data['Type of Diabetes'] == 'T2DM']['BMI (kg/m2)']

# Performing ANOVA again
anova_redo_results = f_oneway(t1dm_bmi, t2dm_bmi)
anova_redo_results
```

```python
from scipy.stats import ttest_ind, chi2_contingency

# T-test on BMI between Types of Diabetes
t1dm_bmi = data[data['Type of Diabetes'] == 'T1DM']['BMI (kg/m2)']
t2dm_bmi = data[data['Type of Diabetes'] == 'T2DM']['BMI (kg/m2)']
t_stat, p_val = ttest_ind(t1dm_bmi, t2dm_bmi)
print(f'T-test results: T-statistic = {t_stat}, p-value = {p_val}')

# Chi-Square Test for association between Gender and Type of Diabetes
contingency_table = pd.crosstab(data['Gender (Female=1, Male=2)'], data['Type of Diabetes'])
chi_stat, chi_p, dof, expected = chi2_contingency(contingency_table)
print(f'Chi-square test results: Chi-stat = {chi_stat}, p-value = {chi_p}')
```

```python
from scipy.stats import ttest_ind

# Analyze impact of smoking on BMI
smokers = data[data['Smoking History (pack year)'] > 0]['BMI (kg/m2)']
nonsmokers = data[data['Smoking History (pack year)'] == 0]['BMI (kg/m2)']
t_stat, p_val = ttest_ind(smokers.dropna(), nonsmokers.dropna())
print(f'Impact of Smoking on BMI: T-statistic = {t_stat}, p-value = {p_val}')

# Analyze impact of alcohol on blood glucose levels
drinkers = data[data['Alcohol Drinking History (drinker/non-drinker)'] == 'drinker']['BMI (kg/m2)']
non_drinkers = data[data['Alcohol Drinking History (drinker/non-drinker)'] == 'non-drinker']['BMI (kg/m2)']
t_stat, p_val = ttest_ind(drinkers.dropna(), non_drinkers.dropna())
print(f'Impact of Alcohol on BMI: T-statistic = {t_stat}, p-value = {p_val}')
```

```python
# Compare blood glucose levels between Type 1 and Type 2 diabetes
t1_glucose = data[data['Type of Diabetes'] == 'T1DM']['Merge 2-hour Postprandial insulin (pmol/L)']
t2_glucose = data[data['Type of Diabetes'] == 'T2DM']['Merge 2-hour Postprandial insulin (pmol/L)']
t_stat, p_val = ttest_ind(t1_glucose.dropna(), t2_glucose.dropna())
print(f'Blood Glucose Levels: T1DM vs T2DM: T-statistic = {t_stat}, p-value = {p_val}')

# Explore complications frequency in Type 1 vs Type 2 diabetes
complications_t1 = data[data['Type of Diabetes'] == 'T1DM']['Acute Diabetic Complications'].value_counts(normalize=True)
complications_t2 = data[data['Type of Diabetes'] == 'T2DM']['Acute Diabetic Complications'].value_counts(normalize=True)
print(f'Complications in T1DM: {complications_t1}')
print(f'Complications in T2DM: {complications_t2}')
```
