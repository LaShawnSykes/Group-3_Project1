```python
# Calculate the mean of 'Duration of Diabetes (years)' for the rows with the condition 'T1DM'
mean_duration_t1dm = df3[df3['Type of Diabetes'] == 'T1DM']['Duration of diabetes (years)'].mean()

# Insert the derived mean into the null values of 'Duration of Diabetes' for rows with the condition 'T1DM'
df3.loc[df3['Type of Diabetes'] == 'T1DM', 'Duration of diabetes (years)'].fillna(mean_duration_t1dm, inplace=True)

# Display the 'Duration of Diabetes (years)' column
print(df3['Duration of diabetes (years)'])
df3.head(17)
```
